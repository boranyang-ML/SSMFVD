# Process the trained weights from Training stage 1

```python
# model path (get from training stage 1)
path = "../model-9804.pth"
m1 = torch.load(path, map_location=torch.device('cpu') )
```

```python
for i in m1.keys():
    print(i)
```

```python
for i in m1["model"].keys():
    print(i)
```

```python
# rename, so the weights get from training stage 1 can be used for training stage 2
for i in list(m1["model"].keys()):
    if i.startswith("base_network."):
        new_key = i.split("base_network.")[1]
        m1["model"][new_key] = m1["model"][i]
        del m1["model"][i]
        
for i in m1["model"].keys():
    print(i)
```

```python
torch.save(m1["model"], "../P2K-nobottle-Ktarget-9804-supervised.pth")
```

```python
# remove num_batches_tracked
for i in list(m1["model"].keys()):
    if i.endswith("num_batches_tracked"):
        del m1["model"][i]
        
for i in m1["model"].keys():
    print(i)
```

```python
torch.save(m1["model"], "../P2K-nobottle-Ktarget-9804-supervised-Notrack.pth")
```
